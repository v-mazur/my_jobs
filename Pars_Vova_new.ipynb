```python
# Импортируем нужные библиотеки
import pandas as pd
import requests 
from bs4 import BeautifulSoup as bs
```

```python
# Создаем ссылку
url = 'https://www.imdb.com/search/title/?release_date=2010-01-01,2022-12-31'
```

```python
# Считываем сайт
request = requests.get(url)
```

```python
# Проверям, что считка прошла успешна
request
```

```python
# Для поиска классов
print(request.text)
```

```python
soup = bs(request.text, "html.parser")
```

```python
# Считаем названия фильмов
names = []
name_film = soup.find_all('h3', class_='lister-item-header') 
for name in name_film:
    names.append(name.text)
```

```python
names # Видно, что зацепились сразу и года. Их разделим уже в датафрэйме
```

```python
# Считаем теперь рейтинг
reiting = []
reiting_info = soup.find_all('span', class_='value')
for info in reiting_info:
    reiting.append(info.text.strip())
```

```python
reiting
```

```python
# Считаем теперь жанр
janre = []
janre_info = soup.find_all('span', class_='genre')
for info_jan in janre_info:
    janre.append(info_jan.text)
```

```python
janre
```

```python
# Считаем теперь режиссеров 
regist = []
regist_info = soup.find_all('p', class_='')
for info_reg in regist_info:
    regist.append(info_reg.text)
```

```python
regist # Видим, что считатлось сразу и инфа о звездах, разделим их в полученном датафрэйме
```
