# Спарсим сайт с фильмами сразу по нескольким страницам

```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
resultat = {'names_film': [], 'raiting': [], 'janre': [], 'director': []}

for i in range (1, 153, 50): # записываем сколько страниц будет считывать (в данном случае 3)
    urles = 'https://www.imdb.com/search/title/?release_date=2010-01-01,2022-12-31&start=' + str(i) # прибавляем страницы
    request = requests.get(urles)
    soup = bs(request.text, "html.parser")
    
    name_film = soup.find_all('h3', class_='lister-item-header') # считываем наименования и записываем в словарь
    for name in name_film:
        resultat['names_film'].append(name.text.strip())


    reiting_info = soup.find_all('span', class_='value') # считываем рейтинг и записываем в словарь
    for reiting in reiting_info:
        resultat['raiting'].append(reiting.text.strip())
        
        
    janre_info = soup.find_all('span', class_='genre') # считываем жанр и записываем в словарь
    for info_jan in janre_info:
        resultat['janre'].append(info_jan.text.strip())
        
    regist_info = soup.find_all('p', class_='') # считываем режиссера/актера и записываем в словарь
    for info_reg in regist_info:
        resultat['director'].append(info_reg.text.strip())
```

```python
df = pd.DataFrame.from_dict(resultat, orient='index') # записываем все в датафрэйм
df = df.transpose()
```

```python
df # видно, что добавились года в наименования, а с режиссерами и звезды
```

```python
# Обработаем таблицу и приведем к человескому виду

df['year'] = df.names_film.apply(lambda x: x.split('\n')[2])
df['names_film'] = df.names_film.apply(lambda x: x.split('\n')[1])
df['Режиссер'] = df.director.apply(lambda x: x.split('Stars')[0])
df['Режиссер'] = df['Режиссер'].str.replace('Director', '')
df['Режиссер'] = df['Режиссер'].str.replace('\n', '')
df['Режиссер'] = df['Режиссер'].str.replace(':', '')
df['Режиссер'] = df['Режиссер'].str.replace('s', '', 1)
df['Актеры'] = df.director.apply(lambda x: x.split('Stars:')[1])
df['Актеры'] = df['Актеры'].str.replace('\n', '')
df = df.drop('director', axis=1)
df = df.rename(columns={'names_film':'Наименование фильма', 'raiting':'Рейтинг', 'janre' : 'Жанр', 'year' : 'Год'})
```

```python
# Переведем колонку 'Рейтинг' в тип данных float
df['Рейтинг'] = df['Рейтинг'].astype(float)
```
