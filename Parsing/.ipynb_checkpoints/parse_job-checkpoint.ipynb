# Парсинг сайта предложений работы

```python
# Запишем данные (данные работы в Одессе)

import requests
import pandas as pd
from bs4 import BeautifulSoup as bs

# Считываем данные с сайта ()
url = 'https://www.work.ua/ru/jobs-odesa/?page=1'
request = requests.get (url)
```

```python
# Проверяем, что все работает (Response [200])
request
```

```python
# Найдем нужные нам для парсинга классы
print(request.text)
```

```python
# Начинаем парсинг
res = {'job': [], 'info': []}
soup = bs(request.text, "html.parser")
vacancies_names = soup.find_all('h2', class_='') # Найдем наименования вакансий (класс и h2 нашли в requests.text)
for name in vacancies_names:
    res['job'].append(name.text.strip()) # Запишем результат в переменную (strip() - чтобы дропнуть /n)
vacancies_info = soup.find_all('p', class_='overflow text-muted add-top-sm cut-bottom') # Описание вакансий
for info in vacancies_info:
    res['info'].append(info.text.strip()) # Запишем результаты в переменную

df = pd.DataFrame.from_dict(res, orient='index') # Используем orient='index', чтобы он не ругался на неравные данные в колонках
df = df.transpose()
df= df.iloc[0:14]# Уберем ненужные строки 
```

### Теперь, когда все работает, запишем все в одну функцию с возможностью менять страницы

```python
url = 'https://www.work.ua/ru/jobs-odesa/?page=1'
def parcing (url):
    #page = int(input('Введите номер страницы: '))
    request = requests.get(url)
    res = {'job': [], 'info': []}
    soup = bs(request.text, "html.parser")
    vacancies_names = soup.find_all('h2', class_='') # Найдем наименования вакансий (класс и h2 нашли в requests.text)
    for name in vacancies_names:
        res['job'].append(name.text.strip()) # Запишем результат в переменную (strip() - чтобы дропнуть /n)
    vacancies_info = soup.find_all('p', class_='overflow text-muted add-top-sm cut-bottom') # Описание вакансий
    for info in vacancies_info:
        res['info'].append(info.text.strip()) # Запишем результаты в переменную

    df = pd.DataFrame.from_dict(res, orient='index') # Используем orient='index', чтобы он не ругался на неравные данные в колонках
    df = df.transpose()
    df= df.iloc[0:14]# Уберем ненужные строки
    
    return df
```

```python
parcing(url)
```
